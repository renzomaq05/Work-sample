```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

datos = pd.read_csv(r"Internal Fraud Dataset.csv")
datos_limpios = datos.drop(['accumulated_payments_by_card_for_reject', 'accumulated_amount_by_buyer_for_reject', 'iv_with_selfie_offered_by_ss_done','accumulated_amount_by_card_for_reject','accumulated_payments_by_buyer_for_reject','approved_history_by_min_amount15after_days60','approved_history_by_min_amount15after_days90','tracker_approved_history','approved_history_by_min_amount0after_days90','address_by_min_amount30after_days90','tracker_old_address','address_by_min_amount0after_days60','approved_history_by_min_amount30after_days90'], axis=1)
```

```python
etiqueta = datos_limpios['bin_country_code'].astype('category').cat.categories.tolist()
mapa_remplazo = {'bin_country_code' : {k: v for k,v in zip(etiqueta,list(range(1,len(etiqueta)+1)))}}
```

```python
datos_limpios.replace(mapa_remplazo, inplace=True)
```

## Análisis de correlación de a pares metodo pearson

```python
datos_limpios.corr(method='pearson')
```

```python
dl_corr = datos_limpios.corr('pearson')
mask = np.triu(dl_corr, k=1)
sns.heatmap(dl_corr, cmap= 'YlGnBu', annot=True, fmt=".2f", mask=mask )
```

## Análisis de correlación de a pares metodo spearman

```python
datos_limpios.corr(method='spearman') 
```

```python
dl_corr = datos_limpios.corr('spearman')
mask = np.triu(dl_corr, k=1)
sns.heatmap(dl_corr, cmap= 'YlGnBu', annot=True, fmt=".2f", mask=mask )
```

## Análisis de correlación de a pares metodo kendall

```python
datos_limpios.corr(method='kendall')
```

```python
dl_corr = datos_limpios.corr('kendall')
mask = np.triu(dl_corr, k=1)
sns.heatmap(dl_corr, cmap= 'YlGnBu', annot=True, fmt=".2f", mask=mask )
```
